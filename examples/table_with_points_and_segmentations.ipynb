```python
import numpy as np
from PIL import Image

project_name = "SegmentationImageWithPointsAnd3DPoints"

column_schemas = {
    # "seg": tlc.SegmentationImagePath("seg", {0: "background", 1: "balloon"}),
    "seg": tlc.SegmentationPILImage("seg", {0: "background", 1: "balloon"}),
    "points": tlc.Geometry2DSchema(include_2d_vertices=True),
    "points_3d": tlc.Geometry3DSchema(include_3d_vertices=True),
}

points = tlc.Geometry2DInstances.create_empty(0, 0, 640, 640)
points.add_instance(vertices=np.random.rand(100, 2) * 640)

points_3d = tlc.Geometry3DInstances.create_empty(0, 640, 0, 640, 0, 640)
points_3d.add_instance(vertices=np.random.rand(100, 3) * 640)

def load_img(path):
    return Image.fromarray(np.array(Image.open(path)))

data = {
    "seg": [load_img(img)],
    "points": [points.to_row()],
    "points_3d": [points_3d.to_row()]
}

table_writer = tlc.TableWriter(project_name=project_name, column_schemas=column_schemas, if_exists="rename")
table_writer.add_batch(data)
table = table_writer.finalize()

# table = tlc.Table.from_dict(
#     data,
#     column_schemas,
#     if_exists="rename"
# )
```
