# Train a YOLO model for object detection with oriented bounding boxes

This notebook trains a YOLO model for object detection with oriented bounding boxes on the [HRSC2016-MS](https://github.com/wmchen/HRSC2016-MS) dataset.

![](../images/hrsc2016-ms-run.png)

<!-- Tags: ["oriented-bounding-boxes", "yolo", "training"] -->

We re-use the Tables created in [create-custom-obb-table.ipynb](../1-create-tables/create-custom-obb-table.ipynb).

## Project setup

```python
PROJECT_NAME = "OBB Example"
DATASET_NAME = "HRSC2016-MS"
DOWNLOAD_PATH = ""
```

## Imports

```python
%pip install 3lc-ultralytics
```

```python
import tlc
from tlc_ultralytics import YOLO, Settings
```

## Load tables

```python
train_table = tlc.Table.from_names("train", DATASET_NAME, PROJECT_NAME)
val_table = tlc.Table.from_names("val", DATASET_NAME, PROJECT_NAME)
```

## Train a model

```python
model = YOLO("yolo11n-obb.pt")
settings = Settings(
    project_name=PROJECT_NAME,
    run_name="train-yolon-obb-hrsc2016-ms",
    image_embeddings_dim=2,
)

model.train(
    tables={"train": train_table, "val": val_table},
    settings=settings,
    label_column_name="obb",
    epochs=10,
    workers=0,
)
```
