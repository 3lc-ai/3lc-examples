# Create Table from COCO - Segmentation

This notebook demonstrates how to create a 3LC Table from a COCO-format dataset for **instance segmentation** tasks.

The resulting table will contain an "image" column with the image filenames, with
corresponding segmentation annotations in the "segmentations" column.


## Project setup


```python
PROJECT_NAME = "3LC Tutorials - COCO128"
DATA_PATH = "../../data/"
```

## Install dependencies

## Imports


```python
from pathlib import Path

import tlc
```

## Create Segmentation Table


```python
annotations_file = (Path(DATA_PATH) / "coco128" / "annotations.json").absolute()
image_folder = (Path(DATA_PATH) / "coco128" / "images").absolute()

assert annotations_file.exists()
assert image_folder.exists()
```

```python
seg_table = tlc.Table.from_coco(
    annotations_file=annotations_file,
    image_folder=image_folder,
    table_name="initial-segmentation",
    dataset_name="COCO128",
    project_name=PROJECT_NAME,
    task="segment",
)
```
