# Create Table from YOLO - Instance Segmentation

This notebook demonstrates how to create a 3LC Table from a YOLO-format dataset for **instance segmentation** tasks.

The resulting table will contain an "image" column with the image filenames, with corresponding segmentation annotations in the "segmentations" column.


## Project setup


```python
PROJECT_NAME = "3LC Tutorials - YOLO Segmentation"
DATASET_NAME = "YOLO-Seg-Dataset"
TABLE_NAME = "initial-segmentation"
DATA_PATH = "../../../data/"
```

## Install dependencies

## Imports


```python
from pathlib import Path

import tlc
```

## Create Segmentation Table



```python
dataset_yaml_file = (Path(DATA_PATH) / "yolo" / "simple.yaml").absolute()

assert dataset_yaml_file.exists()

train_table = tlc.Table.from_yolo(
    dataset_yaml_file=str(dataset_yaml_file),
    split="train",
    project_name=PROJECT_NAME,
    dataset_name=DATASET_NAME,
    table_name=TABLE_NAME,
    task="segment",
)
```
